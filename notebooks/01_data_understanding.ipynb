## Automatisches Finden einer Beispieldatei
In diesem Abschnitt wird eine geeignete Beispieldatei aus den BASt‑Rohdaten automatisch ausgewählt und eingelesen.
Das erleichtert die Arbeit, weil man nicht jedes Mal einen Dateinamen manuell angeben muss.

````python
# Pfad zu den 2025-Daten:
# Hier wird festgelegt, in welchem Ordner sich die Rohdaten befinden. 
data_dir = "C:/Users/Marijke/iCloudDrive/Dokumente/AKAD MS Data Science/GitHub/verlagerungsmodell_gueterverkehr/data/raw/zaehlstellen/2025/Roh_2025_A_S/"

# Automatisch eine Beispiel-CSV finden (erste Monatsdatei)
# Die Schleife sucht die erste Datei, die keine Metadaten-Datei ist (daher der Ausschluss
# ["meta" not in f.lower()]. Das Ergebnis wird in example_file gespeichert
example_file = None
for f in os.listdir(data_dir):     # Auflistung aller Dateien im Ordner
    if f.lower().endswith(".csv") and "meta" not in f.lower():
        example_file = f
        break

print("Gefundene Beispieldatei:", example_file) 

# Erzeugung des vollständigen Dateipfades
# [os.path.join] sorgt dafür, dass der Pfad unabhängig vom Betriebssystem korrekt zusammengesetzt wird
path = os.path.join(data_dir, example_file)
````

## Datei einlesen

````python
df = pd.read_csv(path, sep=";", encoding="latin1")
print("Rohdaten geladen:", df.shape)

df.head()
````

# Data Dictionary – BASt Metadaten (Zählstelleninformationen)

Die Metadaten enthalten alle statischen Informationen zu den automatischen Dauerzählstellen.  
Sie verbinden die Rohdaten (Verkehrsmengen) mit geografischen und infrastrukturellen Eigenschaften der Messstellen.

---

## 🧭 Identifikations- und Verwaltungsinformationen

- **Zst** – Zählstellen-ID (eindeutige Kennung der Messstelle; verbindet Rohdaten und Metadaten)  
- **TKNR** – Technische Knotennummer (interne BASt-ID)  
- **Strklas** – Straßenklasse (A = Autobahn, B = Bundesstraße)  
- **Strnum** – Straßennummer (z. B. 23 = A23)  
- **Strname** – Straßenname oder Bezeichnung (falls vorhanden)  
- **Land** – Bundesland (numerischer Code, z. B. 1 = BW, 5 = NRW)  

---

## 🗺️ Geografische Informationen

- **Latitude** – geografische Breite der Zählstelle (WGS84)  
- **Longitude** – geografische Länge der Zählstelle (WGS84)  
- **Abschnitt** – Abschnittsnummer der Straße (falls vorhanden)  
- **Station** – Stationsnummer innerhalb des Abschnitts (falls vorhanden)  

Diese Spalten ermöglichen die Darstellung der Zählstellen auf Karten (z. B. GeoPandas, Folium).

---

## 🚦 Fahrtrichtungs- und Infrastrukturinformationen

- **Fahrtrichtung** – beschreibt die Richtung der Messung (z. B. „Richtung Hamburg“)  
- **Richtung** – numerische Richtungskennung (1 oder 2)  
- **Fahrstreifen** – Anzahl der Fahrstreifen an der Messstelle (falls vorhanden)  
- **Zst_Typ** – Typ der Messstelle (z. B. Induktionsschleife, Radar)  

---

## 🏗️ Weitere Merkmale

- **Inbetriebnahme** – Datum der Inbetriebnahme der Messstelle  
- **Stilllegung** – Datum der Stilllegung (falls nicht mehr aktiv)  
- **Kommentar** – zusätzliche Hinweise der BASt (z. B. Umbauten, Besonderheiten)  

---
